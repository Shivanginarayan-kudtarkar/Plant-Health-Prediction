```python
# Labelling the data
x=df.drop('Plant_Health_Status',axis=1)
y=df['Plant_Health_Status']
```

```python
#splitting the data
from sklearn.model_selection import train_test_split    
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smt=SMOTE()
x_train_sampled,y_train_sampled=smt.fit_resample(x_train,y_train)
x_train_sampled.shape
```

```python
y_train_sampled.shape
```

```python
y_train.value_counts()
```

```python
y_train_sampled.value_counts()
```

# Logistic Regression

```python
# building the model
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
#model training
lr.fit(x_train_sampled,y_train_sampled)
```

```python
#model testing
result=lr.predict(x_test)
```

```python
result[:100]
```

```python
y_test
```

```python
#model evaluation
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
score=accuracy_score(y_test,result)
print("Accuracy Score For Logistic Model is:",score)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cm=confusion_matrix(y_test,result)
plt.figure(figsize=(8,6),dpi=100)
plt.title("Logistic Regression Matrix")
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted value")
plt.ylabel("Actual Value")
plt.show()
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
# model building
dt=DecisionTreeClassifier()
# model training
dt.fit(x_train_sampled,y_train_sampled)
# model testing
dt_result=dt.predict(x_test)
dt_result
```

```python
dt_acc=accuracy_score(y_test,dt_result)
print("Accuracy score for decision tree is:",dt_acc)
```

```python
cm=confusion_matrix(y_test,dt_result)
plt.figure(figsize=(8,6),dpi=100)
plt.title("Decision Tree classifier Matrix")
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("predicted value")
plt.ylabel("Actual Value")
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
# model building
rf=RandomForestClassifier()
# model training    
rf.fit(x_train_sampled,y_train_sampled)
# model testing
rf_result=rf.predict(x_test)    
rf_result
```
